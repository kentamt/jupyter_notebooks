```python
from pprint import pprint
import numpy as np
import matplotlib.pyplot as plt

from pulp import *
import networkx as nx

%matplotlib inline
warnings.simplefilter('ignore')
```

```python
def make_spring_pos(nodes,shift):
    """A function for shifting node positions"""
    return { node:(pos[0]+shift,pos[1]) for node,pos in nx.spring_layout(G.subgraph(nodes)).items() }

def non_zero_index(mat):
    """Return indies of non-zero element as set"""
    return set(zip(*np.where(mat > 0)))
```

## ヒッチコック輸送問題
入力（生産）される数と出力（消費）される数が等しいという条件。各エッジにどれだけ分配すれば輸送効率を最大化できるか。

```python
G = nx.DiGraph()
G.add_nodes_from([1,2,3,4,5]) # you need to set node in order
G.add_weighted_edges_from([
    (1, 3, 6), (1, 4, 7), (1, 5, 10), 
    (2, 3, 8), (2, 4, 5), (2, 5, 6)
    ],weight="weight")

pos = {
    1:(0, -0.3),
    2:(0, 0.3),
    3:(0.5, 1),
    4:(0.5, 0),
    5:(0.5, -1)
}

# draw
fig = plt.figure(figsize=(5, 5))
edge_labels = {(i, j): w['weight'] for i, j, w in G.edges(data=True)}

text = nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
for _,t in text.items():
    t.set_rotation('horizontal')
    
nx.draw_networkx_labels(G, pos=pos)
nx.draw(G, pos=pos, node_size=1000)
```

Fig 1; エッジの数字は輸送コスト. ノード1の生産量200, ノード2の生産量100, ノード3, 4, 5の消費量はそれぞれ90, 130, 50

### 線形計画法で輸送コストの最小値を求める
輸送コストを最小にするためにどう配送すればいいのかを考える。xを各走路お輸送量、cをコスト、sを生産量、dを消費量とする。目的関数及び制約条件は以下の通り。

```python
# variables
x1 = LpVariable("x1", lowBound=0.0)
x2 = LpVariable("x2", lowBound=0.0)
x3 = LpVariable("x3", lowBound=0.0)
x4 = LpVariable("x4", lowBound=0.0)
x5 = LpVariable("x5", lowBound=0.0)
x6 = LpVariable("x6", lowBound=0.0)
c = np.array([6, 8, 7, 5, 10, 6])
s = np.array([200, 100])
d = np.array([90, 130, 50])
x = np.array([x1, x2, x3, x4, x5, x6])

prob = LpProblem(name="Hitchcock", sense=LpMinimize)
# Objective
prob += c.dot(x)
# where
prob += x1 + x3 + x5 <= s[0] # s1
prob += x2 + x4 + x6 <= s[1] # s2
prob += x1 + x2 == d[0] # d1
prob += x3 + x4 == d[1] # d2
prob += x5 + x6 == d[2] # d3
pprint(prob)
```

ソルバーで解を求める。

```python
prob.solve()
print(f"Result: {LpStatus[prob.status]}")
```

```python
optim = value(prob.objective)
print(f"Objective:\n{prob.objective} = {optim}")
for v in prob.variables():
    print(v.name ,"=", value(v))
```

```python
# prepare labels 
edge_labels = {(e[0], e[1]): value(v) for e, v in zip(G.edges, prob.variables())}

# draw
fig = plt.figure(figsize=(5, 5))
text = nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
for _,t in text.items():
    t.set_rotation('horizontal')
nx.draw_networkx_labels(G, pos=pos)
nx.draw(G, pos=pos, node_size=1000)
```

Fig; エッジの数字は流量を表す。このように流せばコストは最小化する。

## ネットワーク輸送問題
途中に経由地を設置する。また、途中の走路に上限がある場合や、走路が使用できなくなった場合についても最適値を求める。

```python
G = nx.DiGraph()
G.add_nodes_from([1,2,3,4,5,6,7,8]) # you need to set node in order
G.add_weighted_edges_from([(1, 3, 3), 
                           (2, 3, 5), (2, 4, 2), 
                           (3, 4, 5), (3, 5, 4), (3, 6, 3), (3, 7, 4),
                           (4, 3, 5), (4, 5, 2), (4, 8, 4),
                           (5, 3, 4), (5, 4, 2), (5, 6, 6), (5, 7, 1), (5, 8, 3)
                          ],
                          weight="weight")

pos = {
    1:(0, 0.3),
    2:(0, -0.3),
    3:(0.5, 1),
    4:(0.5, -1),
    5:(0.65, 0),
    6:(1, 1),
    7:(1, 0),
    8:(1, -1)
}

# draw    
fig = plt.figure(figsize=(5, 4))
edge_labels = {(i, j): w['weight'] for i, j, w in G.edges(data=True)}
text = nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
for _,t in text.items():
    t.set_rotation('horizontal')
nx.draw_networkx_labels(G, pos=pos)
nx.draw(G, pos=pos, node_size=1000)
```

Fig 2; ネットワーク輸送問題。エッジの数字は輸送コスト。 ノード1の生産量200, ノード2の生産量100, ノード3, 4, 5の消費量はそれぞれ90, 130, 50

### 線形計画法で輸送コストの最小値を求める
xを各走路の輸送量、cをコスト、sを生産量、dを消費量とする。目的関数及び制約条件は以下の通り。

```python
# adjacent matrix 
adj_mat = nx.to_numpy_matrix(G)
print("Adjacent Matrix:")
print(adj_mat, end="\n\n")

# set of edges
# edge_set = non_zero_index(adj_mat)
edge_set = G.edges() # use API

print("Set of edge:")
pprint(edge_set)
```

```python
x_ = {}
for edge in sorted(edge_set, key=lambda x:x[0]):
    suffix = str(edge[0]) + str(edge[1])
    x = LpVariable("x_"+suffix, lowBound=0.0)
    x_[suffix] = x
    
x = np.array(list(x_.values()))
print("Variable dictionary")
print(x_)
print("Variable array")
print(x)
```

```python
c = []
for edge in sorted(edge_set, key=lambda x:x[0]):
    i, j = edge
    i = i - 1
    j = j - 1
    elem = adj_mat[i, j]
    c.append(elem)
c = np.array(c)

print("Weights of edges:")
print(c)
```

```python
s = np.array([200, 100])
d = np.array([90, 130, 50])

prob = LpProblem(name="Network", sense=LpMinimize)

# Objective
prob += c.dot(x)
# where
prob += x_["13"] <= s[0]
prob += x_["23"] + x_["24"] <= s[1]
prob += x_["13"] + x_["23"] + x_["43"] + x_["53"] == x_["34"] + x_["35"] + x_["36"] + x_["37"]
prob += x_["24"] + x_["34"] + x_["54"] == x_["43"] + x_["45"] + x_["48"]
prob += x_["35"] + x_["45"] == x_["53"] + x_["54"] + x_["56"] + x_["57"] + x_["58"]
prob += x_["36"] + x_["56"] == d[0]
prob += x_["37"] + x_["57"] == d[1]
prob += x_["48"] + x_["58"] == d[2]

pprint(prob)
```

```python
# limit ==> 
sub_set = {(3, 4), (3, 5), (4, 3), (4, 5), (5, 3), (5, 4)}
u = 40
for s in sub_set:
    key = str(s[0]) + str(s[1])
    prob += x_[key] <= u
    
# disable some edges
sub_set = {(4, 5), (5, 4)}
u = 0
for s in sub_set:
    key = str(s[0]) + str(s[1])
    prob += x_[key] <= u
```

```python
prob.solve()
print(f"Result: {LpStatus[prob.status]}")
```

```python
optim = value(prob.objective)
print(f"Objective:\n{prob.objective} = {optim}")
for v in prob.variables():
    print(v.name ,"=", value(v))
```

```python
G_ = nx.DiGraph()
G_.add_nodes_from([1,2,3,4,5,6,7,8]) # you need to set node in order
G_.add_weighted_edges_from([(e[0], e[1], value(v)) for e, v in zip(G.edges, prob.variables()) if value(v) != 0],weight="x")

# prepare labels 
edge_labels = {(e[0], e[1]): value(v) for e, v in zip(G.edges, prob.variables()) if value(v) != 0}

# draw
nx.draw_networkx_labels(G_, pos=pos)
text = nx.draw_networkx_edge_labels(G_, pos, edge_labels=edge_labels)
for _,t in text.items():
    t.set_rotation('horizontal')

nx.draw(G_, pos=pos, node_size=1000)
```

Fig 3; コスト最小となる輸送路とその交通量

### 動的計画法で最短経路を求める
輸送コストが最小となる経路を計算する

```python
shortest_path_nodes = nx.algorithms.shortest_paths.weighted.dijkstra_path(G, 2, 7, weight='weight')
shortest_edge_set = {(shortest_path_nodes[i], shortest_path_nodes[i+1]) for i in range(len(shortest_path_nodes)-1)}
```

```python
G_ = nx.DiGraph()
G_.add_nodes_from([1,2,3,4,5,6,7,8]) # you need to set node in order
G_.add_weighted_edges_from([(e[0], e[1], value(v)) for e, v in zip(G.edges, prob.variables()) if value(v) != 0],weight="x")

# prepare labels 
edge_labels = {(e[0], e[1]): value(v) for e, v in zip(G.edges, prob.variables()) if value(v) != 0}

# draw
nx.draw_networkx_labels(G_, pos=pos)
text = nx.draw_networkx_edge_labels(G_, pos, edge_labels=edge_labels)
for _,t in text.items():
    t.set_rotation('horizontal')

nx.draw_networkx_edges(G,pos, edgelist=shortest_edge_set, width=5,edge_color='b')    
nx.draw(G_, pos=pos, node_size=1000)
```

Fig 3; 輸送コスト最小で2から7へ向かう経路
